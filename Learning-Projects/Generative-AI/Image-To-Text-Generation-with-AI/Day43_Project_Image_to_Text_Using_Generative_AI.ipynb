# Import All Required Libraries

````python
import google.generativeai as genai
import PIL.Image
import textwrap
from IPython.display import Markdown
````

# Define Markdown Output Formatter

This helps make the AI’s response easier to read.

````python
def to_markdown(text):
    text = text.replace('-', '•')  # Optional: change bullets
    return Markdown(textwrap.indent(text, '> ', predicate=lambda _: True))
````

# Load the Image
Make sure the image path is correct on your PC.

````python
# This loads and shows your image.
img_path = r"C:\Users\Lenovo\Downloads\1.png"  # Change this path if needed
img = PIL.Image.open(img_path)
img
````

# Generate Text from Image using Gemini 2.5 Flash Model

- generate_content(img) sends the image to Gemini

- Gemini returns a description or analysis of what's in the image

````python
model = genai.GenerativeModel('models/gemini-2.5-flash-preview-05-20')  # image understanding model
response = model.generate_content(img)
to_markdown(response.text)
````

# Use Gemini 2.5 Thinking Model on the Same Image

This model is better for deep reasoning.

````python
model = genai.GenerativeModel('models/gemini-2.5-flash-preview-04-17-thinking')
response = model.generate_content(img)
to_markdown(response.text)
````
